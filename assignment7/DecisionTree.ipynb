```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.tree import plot_tree
```

```python
data=pd.read_csv('/content/Admission_Predict_Ver1.1.csv')
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['TOEFL Score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of TOEFL Scores', fontsize=16)
plt.xlabel('TOEFL Score', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.gca().spines[['top', 'right']].set_visible(False)

plt.show()
```

```python
label_encoder = LabelEncoder()
data['Research'] = label_encoder.fit_transform(data['Research'])
```

```python
data.columns = data.columns.str.strip()
```

```python
x = data[['GRE Score','CGPA']]
y = data['Chance of Admit']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50)
y_test = np.array(y_test)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(x_train['GRE Score'], x_train['CGPA'], c=y_train, cmap='BuPu', marker='o', label='Train Data')
plt.xlabel('GRE Score')
plt.ylabel('CGPA')
plt.title('Train Data')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(x_test['GRE Score'], x_test['CGPA'], c=y_test, cmap='BuPu', marker='o', label='Test Data')
plt.xlabel('GRE Score')
plt.ylabel('CGPA')
plt.title('Test Data')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
model = DecisionTreeRegressor(max_depth=3, random_state=50)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred = np.array(y_pred)
print(y_pred.shape)
```

```python
plt.figure(figsize=(12, 8))
plot_tree(model, filled=True, feature_names=x.columns, class_names=['Not Admitted', 'Admitted'])
plt.title("Decision Tree Max Depth = 3")
plt.show()
```

```python
print("Mean Square Error:  ",metrics.mean_squared_error(y_test,y_pred))
print("Mean Absolute Error: ", metrics.mean_absolute_error(y_test,y_pred))
print("R-Sqaure : ", metrics.r2_score(y_test,y_pred))
```
