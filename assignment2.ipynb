```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#importing the data
admissions = pd.read_csv("/content/admission - admission.csv")
admissions.head()
```

```python
# Scatter plot for GRE Score vs Chance od Admissions
admissions.hist(bins=30, color='r', align='right')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.suptitle("Histogram for Admission Dataset")
plt.tight_layout()
plt.show()
```

```python
# Dropping duplicates
admissions = admissions.drop_duplicates()

# Dropping rows with missing values
admissions.dropna(inplace=True)

# Displaying the cleaned DataFrame
print(admissions.head())
```
