```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing data
sales = pd.read_excel("/content/supermarket_sales.xlsx")
sales.head()
```

```python
sales.info()
```

```python
#ploting a bar chart
plt.figure(figsize=(8,5))
ax=sns.countplot(x="Product line", data=sales, palette='GnBu')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right", fontsize=10)
```

```python
#Scatter plot for unit price and gross income
plt.scatter(sales['Unit price'], sales['gross income'])
sns.set_style("darkgrid", {'grid.color': '.5'})
sns.set_style("whitegrid")
sns.despine()
plt.show()
```

```python
sns.boxplot(x=sales['Payment'], y=sales['Quantity'],  palette='PuRd', width=0.5)
```

```python
x=sales['Unit price']
plt.hist(x,bins=100, align="mid", histtype="bar", color='c')
plt.xlabel('Unit price')
plt.ylabel('frequency')
plt.title('Histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.countplot(x="Payment", data=sales, order=sales["Payment"].value_counts().index)
plt.title("Distribution of Payment")
plt.xlabel("Payment")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Check column names and remove leading/trailing whitespaces
sales.columns = sales.columns.str.strip()
print(sales.columns)

# Selecting the specified columns
selected_columns = sales[['Unit price', 'Quantity', 'Total', 'Rating']]

# Calculating the correlation matrix
correlation_matrix = selected_columns.corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Selected Attributes')
plt.show()

```
